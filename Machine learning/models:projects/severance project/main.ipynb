```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

df = pd.read_csv('severance_pay.csv')

#lets create a few dataframes

#dataframe no gender
df1 = df.drop('Gender', axis = 1)

#df1 modeling
inputs1 = df1.drop('Weeks SP', axis = 1)
inputs1 = sm.add_constant(inputs1)
outputs1 = df1['Weeks SP']

model1 = sm.OLS(outputs1,inputs1).fit()
model1.summary()


```

```python

#dataframe no pay, no gender
df2 = df.drop(['Pay','Gender'],axis = 1)

#df2 modeling
inputs2 = df2.drop('Weeks SP', axis = 1)
inputs2 = sm.add_constant(inputs2)
outputs2 = df2['Weeks SP']

model2 = sm.OLS(outputs2,inputs2).fit()
model2.summary()
```
